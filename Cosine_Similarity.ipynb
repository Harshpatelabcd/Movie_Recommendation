```python
credits = pd.read_csv('credits.csv')
```

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
import ast
def convert1(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['poster_path'])
    return L
```

```python
movies['belongs_to_collection']=movies['belongs_to_collection'].apply(convert1)
```

```python
movies['belongs_to_collection']=movies['belongs_to_collection'].str[0]
```

```python
ratings_with_id.head(1)
```

users who gave minimum 30 ratings to movies...... 

```python
x=ratings_with_id.groupby('userId').count()['rating'] > 20
litrate_user = x[x].index
```

out of all user's who had rated the movies...this will filtered the user's who are litrate....

```python
filtered_rating = ratings_with_id[ratings_with_id['userId'].isin(litrate_user)]
```

movies that have minimum 30 ratings....

```python
y = filtered_rating.groupby('original_title').count()['rating']>=20
famous_books = y[y].index
```

out of all movies.. this will filtered the movies that have 30 or more ratings...

```python
final_ratings = filtered_rating[filtered_rating['original_title'].isin(famous_books)]
```

```python
final_ratings.shape
```

creating a pivot tables....

```python
pt = final_ratings.pivot_table(index='original_title',columns='userId',values='rating') 
pt
```

```python
pt.fillna(0,inplace=True)
pt
```

Identifying Uclidian distance of one movies with other 108 movies..... 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

Function that identifies the 6 movies related to movie title which uclidian distance is nearer....

```python
index2 = np.where(pt.index=="X-Men Origins: Wolverine")[0][0]
index2

similar_items1 = sorted(list(enumerate(similarity_scores[index2])),key=lambda x:x[1],reverse=True)[1:6]
similar_items1
```

```python
def recommend(movie_name):
    #index fetch
    index1 = np.where(pt.index==movie_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index1])),key=lambda x:x[1],reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = movies[movies['original_title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('original_title')['original_title'].values))
        item.extend(list(temp_df.drop_duplicates('original_title')['crew'].values))
        item.extend(list(temp_df.drop_duplicates('original_title')['belongs_to_collection'].values))
        
        data.append(item)
        
    return data
```

```python
recommend('2001: A Space Odyssey')
```

```python
import pickle
pickle.dump(pt,open('pt.pkl','wb'))
pickle.dump(movies,open('movies.pkl','wb'))
pickle.dump(similarity_scores,open('similarity_scores.pkl','wb'))
```
